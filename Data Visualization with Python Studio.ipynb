```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
import numpy as np

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
df.sort_values(by='year', inplace = True)

```

With your dataframe set up, time to start creating visualizations!

```python
plt.scatter('ratings_count','text_reviews_count',data= df, s=200, linewidth= 1, edgecolor= 'blue',color= 'blue', alpha=0.6)
```

```python
# Use your Python skills to create a visualization here.
sns.scatterplot('ratings_count', 'text_reviews_count', data= df, s=200, linewidth= 1, edgecolor= 'blue',color= 'blue', alpha=0.6)
plt.title('Text ratings count versus reviews count', color= 'black', fontsize= 18)
plt.savefig('scatter.png')

plt.show()
```

```python
year_counts= df['year'].value_counts()
color = ['blue','red']
#years = df.groupby('year')['year'].value_counts()
year_df = df['year'].value_counts().rename_axis('year').reset_index(name='counts')
year_df.sort_values(by='year', inplace = True)
year_df.tail()
year_df.plot(kind = 'area', stacked = False, color = color)
```
